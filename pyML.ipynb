De asemenea, cunoaștem diagnosticul fiecărui pacient: sănătos sau bolnav.
Scopul este să construim un algoritm care primind caracteristicile fiecărui pacient (vârstă, sex, bmi, bp și rezultatele analizelor de sânge), să ne poată spune dacă pacientul este unul sănătos sau dacă suferă de diabet.

```python
#@markdown Rulați această celulă pentru a încărca și vizualiza datele (de asemenea, se vor importa bibliotecile necesare și se vor defini funcții auxiliare).
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

## definire functii utile

# util pentru a putea afisa linia de demarcatie si cele 2 zone rezultate
def make_grid_predictions(df, column_1, column_2, a, b, c, type, step):
  if type == 'univariate':
    min_x = 0
    max_x = len(df)
  else:
    min_x = np.min(df[column_1])
    max_x = np.max(df[column_1])

  min_y = np.min(df[column_2])
  max_y = np.max(df[column_2])

  x = np.arange(min_x-0.5, max_x+0.5, step)
  y = np.arange(min_y-0.5, max_y+0.5, step)
  xx, yy = np.meshgrid(x,y)
  grid_pred = (a*xx+b*yy+c) >= 0

  return xx, yy, grid_pred

# util pentru a putea afisa linia de demarcatie si cele 2 zone rezultate
def make_grid_predictions_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5, step):
  min_x = np.min(df[column_1])
  max_x = np.max(df[column_1])

  min_y = np.min(df[column_2])
  max_y = np.max(df[column_2])

  x = np.arange(min_x-0.5, max_x+0.5, step)
  y = np.arange(min_y-0.5, max_y+0.5, step)
  xx, yy = np.meshgrid(x,y)
  grid_pred = (w0 + w1*xx + w2*yy + w3*xx*xx + w4*xx*yy + w5*yy*yy) >=0

  return xx, yy, grid_pred

# functie care stabileste ce categorie de pacienti avem de partea pozitiva a
#    liniei de demarcatie
# utila doar pentru exemplul curent in care nu stabilim apriori acest lucru
def establish_above_boundary_class(df, column_1, column_2, a, b, c):

  xx = df[column_1]
  yy = df[column_2]
  prediction = (a*xx + b*yy + c) >= 0

  annotation = df['diagnostic'] == 'bolnav'
  p_sick_pos = (  np.sum(prediction * annotation) +
                np.sum((~prediction) * (~annotation))) / len(df)
  p_healthy_pos = (  np.sum((~prediction) * annotation) +
                np.sum(prediction * (~annotation))) / len(df)
  if p_sick_pos > p_healthy_pos:
    above_boundary_class = 'bolnav'
    below_boundary_class = 'sanatos'
  else:
    above_boundary_class = 'sanatos'
    below_boundary_class = 'bolnav'

  return above_boundary_class, below_boundary_class

# afiseaza rezultatele statistice
def display_metrics(df, column_1, column_2, a, b, c):
  above_boundary_class, below_boundary_class =\
    establish_above_boundary_class(df, column_1, column_2, a, b, c)

  prediction = (a*df[column_1]+b*df[column_2]+c) >= 0
  annotation = df['diagnostic'] == above_boundary_class
  p_above = np.sum(prediction * annotation) / np.sum(annotation)
  p_below = np.sum((~prediction) * (~annotation)) / np.sum(~annotation)
  acc = (np.sum(prediction * annotation) + np.sum((~prediction) * (~annotation)))/ len(df)
  #print('Categorie de pacienți deasupra linie de demarcație: %s'%above_boundary_class)
  #print('-------------------------------------------------------------')
  if above_boundary_class == 'bolnav':
    print("%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați"%(p_above*100, above_boundary_class))
    print('%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați'%(p_below*100, below_boundary_class))
  else:
    print('%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați'%(p_below*100, below_boundary_class))
    print("%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați"%(p_above*100, above_boundary_class))
  print('%4.2f %% din pacienți au fost diagnosticați corect'%(acc*100))
  print('-------------------------------------------------------------')

# functie care stabileste ce categorie de pacienti avem de partea pozitiva a
#    liniei de demarcatie
# utila doar pentru exemplul curent in care nu stabilim apriori acest lucru
def establish_above_boundary_class_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5):

  xx = df[column_1]
  yy = df[column_2]
  prediction = (w0 + w1*xx + w2*yy + w3*xx*xx + w4*xx*yy + w5*yy*yy) >=0

  annotation = df['diagnostic'] == 'bolnav'
  p_sick_pos = (  np.sum(prediction * annotation) +
                np.sum((~prediction) * (~annotation))) / len(df)
  p_healthy_pos = (  np.sum((~prediction) * annotation) +
                np.sum(prediction * (~annotation))) / len(df)
  if p_sick_pos > p_healthy_pos:
    above_boundary_class = 'bolnav'
    below_boundary_class = 'sanatos'
  else:
    above_boundary_class = 'sanatos'
    below_boundary_class = 'bolnav'

  return above_boundary_class, below_boundary_class

# afiseaza rezultatele statistice
def display_metrics_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5):
  above_boundary_class, below_boundary_class =\
    establish_above_boundary_class_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5)

  xx = df[column_1]
  yy = df[column_2]

  prediction = (w0 + w1*xx + w2*yy + w3*xx*xx + w4*xx*yy + w5*yy*yy) >=0
  annotation = df['diagnostic'] == above_boundary_class
  p_above = np.sum(prediction * annotation) / np.sum(annotation)
  p_below = np.sum((~prediction) * (~annotation)) / np.sum(~annotation)
  acc = (np.sum(prediction * annotation) + np.sum((~prediction) * (~annotation)))/ len(df)
  #print('Categorie de pacienți deasupra linie de demarcație: %s'%above_boundary_class)
  #print('-------------------------------------------------------------')
  if above_boundary_class == 'bolnav':
    print("%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați"%(p_above*100, above_boundary_class))
    print('%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați'%(p_below*100, below_boundary_class))
  else:
    print('%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați'%(p_below*100, below_boundary_class))
    print("%4.2f %% din pacienții cu eticheta \'%s\' au fost corect identificați"%(p_above*100, above_boundary_class))
  print('%4.2f %% din pacienți au fost diagnosticați corect'%(acc*100))
  print('-------------------------------------------------------------')


## descarcare set de date

!wget https://drive.google.com/uc?id=1vIofujKu2bAgpS9g6Kw4dIIV4Ald0GTX -q -O diabetes_data.csv
import pandas as pd
df = pd.read_csv('diabetes_data.csv')

df.head()
```

Pentru a începe să construim un astfel de algoritm, putem încerca să stabilim câte o valoare limită pentru fiecare caracteristică, în funcție de care să alegem un anumit diagnostic (pacienții ale căror valori sunt mai mari decât această valoare limită primesc un diagnostic, iar ceilalți pacienți, cu valori mai mici, primesc celălalt diagnostic).

```python
#@markdown Rulați această celulă pentru a testa reguli simple de diagnosticare.

# functie care regnereaza graficul in functie de noile valori setate
def refresh_plot(th, column):
  # util pentru a putea afisa linia de demarcatie si cele 2 zone rezultate
  xx, yy, grid_pred =\
    make_grid_predictions(df, column, column, 0, 1, -th, 'univariate', 0.1)

  # afisare puncte din setul de date
  sns.scatterplot(data=df, x=df.index, y=column, hue='diagnostic')
  # afisare linie de demarcatie si cele 2 zone rezultate
  plt.contourf(xx, yy, grid_pred, alpha=0.1)

  print('-------------------------------------------------------------')
  print('Valoare limită: %4.2f'%th)
  print('-------------------------------------------------------------')

  # afiseaza rezultatele statistice
  display_metrics(df, column, column, 0, 1, -th)

_ = widgets.interact(refresh_plot,
                     th=widgets.FloatSlider(min=-4.5, max=4.5, step=0.2, value=0, description='valoare limită'),
                     column= widgets.Dropdown(options=df.columns[:-1], description='caracteristică'))
```

Cum citim graficul?
*   în funcție de valoarea limită aleasă, construim o linie de demarcație
*   pacienții reprezentați de puncte aflate deasupra liniei primesc un diagnostic, iar cei aflați în zona de sub linie primesc un alt diagnostic

Puteți glisa slider-ul pentru a alege o altă valoare limită. Similar cu exemplul anterior, puteți folosi dropdown-ul pentru a vizualiza o altă caracteristică. Ideal, ne dorim ca toate punctele protocalii să fie de o parte a liniei de demarcație, iar toate punctele albastre să fie de cealaltă parte a liniei. Dar, pentru problema curentă acest comportament nu poate fi obținut folosind o astfel de regulă simplă.

Conform regulii noastre, unii pacienți vor primi un diagnostic corect, in timp ce alți pacienți vor primi un diagnostic greșit. Deasupra graficului puteți observa procentele de pacienți care au fost diagnosticați corect. (pentru această statistică am considerat cea mai favorabilă alocare a regiunilor de sub si deasupra liniei de demarcație în raport cu diagnosticele)

**Interpretare grafică**

În momentul în care alegem o astfel de valoare limită, practic definim o dreaptă ce separă cele două categorii de pacienți.

Considerând ecuația dreptei:

$d: ax+by+c=0$

pentru fiecare valoare limită $p$, definim o dreaptă cu $a=0, b=1, c=-p$ :

$d: y-p=0$

<font color='AAAAFF'>**Q: Puteți da exemplu de o regulă pe care o putem folosi în această situație?**

---

Putem încerca să determinăm o dreaptă ce separă cele două categorii de pacienți. Spre deosebire de cazul anterior, dreapta va fi determinată de ambele caracteristici.

Plecând de la ecuația $d: ax+by+c=0$, identificarea dreptei ce separă cele două categorii de pacienți este echivalentă cu determinarea parametrilor  $a,b,c$. Avem un algoritm mai complex, ce implică determinarea a 3 parametrii, spre deosebire de cazul anterior.

Putem încerca să determinăm manual acești parametrii.

```python
#@markdown Rulați această celulă pentru a testa linii de demarcație definite de perechi de caracteristici.

# functie care regnereaza graficul in functie de noile valori setate
def refresh_plot(a, b, c, column_1, column_2):
  # util pentru a putea afisa linia de demarcatie si cele 2 zone rezultate
  xx, yy, grid_pred =\
    make_grid_predictions(df, column_1, column_2, a, b, c, 'bivariate', 0.01)

  # afisare puncte din setul de date
  sns.scatterplot(data=df, x=column_1, y=column_2, hue='diagnostic')
  # afisare linie de demarcatie si cele 2 zone rezultate
  plt.contourf(xx, yy, grid_pred, alpha=0.1)
  print('-------------------------------------------------------------')
  print('(a,b,c): (%4.2f, %4.2f, %4.2f)'%(a,b,c))
  print('-------------------------------------------------------------')
  # afiseaza rezultatele statistice
  display_metrics(df, column_1, column_2, a, b, c)

_ = widgets.interact(refresh_plot,
                     a= widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=2, description='a'),
                     b= widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=-1, description='b'),
                     c= widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=0, description='c'),
                     column_1= widgets.Dropdown(options=df.columns[:-1], value= 'GLU', description='caracteristică 1'),
                     column_2= widgets.Dropdown(options=df.columns[:-1], value='LTG', description='caracteristică 2'))
```

Noul algoritm ține cont de perechi de caracteristici ale pacienților, dar totuși nu reușim să identificăm exact cele două categorii de pacienți.

<font color='AAAAFF'>**Q: Cum credeți că am putea îmbunătății deciziile? Putem oare să construim reguli mai complexe?**

---

Putem alege să construim linia de demarcație folosind o funcție polinomială.

Dacă avem o singură caracteristică $x$, funcția polinomială de gran $n$ care ne separă cele două categorii de pacienți ar avea forma:

$f(x) = w_0 + w_1*x + w_2*x^2 + ... + w_n*x^n$

unde $n$ este gradul polinomului, iar $w_0, w_1, ... w_n$ sunt parametrii acestuia (echivalentul parametrilor $a,b,c$ pentru dreapta de mai sus)

Dacă avem 2 caracteristici $x$ și $y$, funcția polinomială de grad 2 care ne poate ajuta să separăm cele două categorii de pacienți va fi de forma:

$f(x,y)= w_0 + w_1*x + w_2*y + w_3*x^2 + w_4*x*y+ w_5*y^2$

```python
#@markdown Rulați această celulă pentru a construi o funcție polinomială de demarcație.
def refresh_plot(w0, w1, w2, w3, w4, w5, column_1, column_2):
  # util pentru a putea afisa linia de demarcatie si cele 2 zone rezultate
  xx, yy, grid_pred =\
    make_grid_predictions_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5, 0.01)

  # afisare puncte din setul de date
  sns.scatterplot(data=df, x=column_1, y=column_2, hue='diagnostic')
  # afisare linie de demarcatie si cele 2 zone rezultate
  plt.contourf(xx, yy, grid_pred, alpha=0.1)
  print('-------------------------------------------------------------')
  print('(w0, w1, w2, w3, w4, w5): (%4.2f, %4.2f, %4.2f,%4.2f, %4.2f, %4.2f)'%(w0, w1, w2, w3, w4, w5))
  print('-------------------------------------------------------------')
  # afiseaza rezultatele statistice
  display_metrics_poly_2(df, column_1, column_2, w0, w1, w2, w3, w4, w5)

_ = widgets.interact(refresh_plot,
                     w0 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=0, description='w_0'),
                     w1 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=2.167, description='w_1'),
                     w2 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=-1.14, description='w_2'),
                     w3 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=0.5, description='w_3'),
                     w4 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=0.048, description='w_4'),
                     w5 = widgets.FloatSlider(min=-4.5, max=4.5, step=0.5, value=-0.007, description='w_5'),
                     column_1= widgets.Dropdown(options=df.columns[:-1], value= 'GLU', description='caracteristică 1'),
                     column_2= widgets.Dropdown(options=df.columns[:-1], value='LTG', description='caracteristică 2'))
```

Acum avem posibilitatea de a construi un model mult mai complex, și în cazul în care identificăm parametrii potriviți putem separa aproape perfect cele două categorii de pacienți (de exemplu pentru perechea de caracteristici 'GLU' și 'LTG'). ###**Determinarea automată a parametrilor funcției polinomiale**

<font color='#AAAAFF'>**Q: Cum putem identifica automat cei mai buni parametrii?**</font>

Putem utiliza un algoritm de învățare automată pentru a determina parametrii optimi. Intuitiv, modul în care sunt calculați parametrii seamănă foarte mult cu modul nostru manual de a identifica acești parametrii:
*   pornim de la un set de parametrii
*   analizăm cât suntem de mulțimiți cu rezultatul generat cu acești parametrii
*   dacă modelul nu este încă optim, încercăm să modificăm parametrii astfel încât să ajungem la o versiune mai bună

Cum exact se ajustează parametrii? https://www.pyml.ro/

```python
#@markdown Rulați această celulă pentru a utiliza un algoritm de învățare automată (Machine Learning) pentru rezolvarea problemei

def refresh_plot(grad, column_1, column_2):
  poly = PolynomialFeatures(grad)
  samples = np.concatenate((np.array(df[column_1])[:,None], np.array(df[column_2])[:,None]), 1)
  labels = df['diagnostic'].tolist()
  labels = [float(label == 'bolnav') for label in labels]
  labels = np.array(labels)

  polynomial_samples = poly.fit_transform(samples)
  model = LogisticRegression()
  model.fit(polynomial_samples, labels)
  pred_labels = model.predict(polynomial_samples)

  min_x = np.min(df[column_1])
  max_x = np.max(df[column_1])
  min_y = np.min(df[column_2])
  max_y = np.max(df[column_2])
  x = np.arange(min_x-0.5, max_x+0.5, 0.01)
  y = np.arange(min_y-0.5, max_y+0.5, 0.01)
  xx, yy = np.meshgrid(x,y)
  xx_ = xx.flatten()[:,None]
  yy_ = yy.flatten()[:,None]
  grid_samples = np.concatenate((xx_, yy_), axis=1)
  polynomial_grid_samples = poly.fit_transform(grid_samples)
  grid_predict = model.predict(polynomial_grid_samples)

  # afisare puncte din setul de date
  sns.scatterplot(data=df, x=column_1, y=column_2, hue='diagnostic')
  # afisare linie de demarcatie si cele 2 zone rezultate
  plt.contourf(xx, yy, grid_predict.reshape(xx.shape), alpha=0.1)
  print('-------------------------------------------------------------')
  print('grad: %d'%grad)
  p_sick = np.sum(labels * pred_labels) / np.sum(labels)
  p_healthy = np.sum((1-labels) * (1-pred_labels)) / np.sum(1-labels)
  acc = (np.sum(labels * pred_labels) + np.sum((1-labels) * (1-pred_labels)))/ len(df)
  print('-------------------------------------------------------------')
  print("%4.2f %% din pacienții cu eticheta \'bolnav' au fost corect identificați"%(p_sick*100))
  print('%4.2f %% din pacienții cu eticheta \'sanatos\' au fost corect identificați'%(p_healthy*100))
  print('%4.2f %% din pacienți au fost diagnosticați corect'%(acc*100))
  print('-------------------------------------------------------------')

_ = widgets.interact(refresh_plot,
                     grad= widgets.IntSlider(min=1, max=10, step=1, value=3, description='grad'),
                     column_1= widgets.Dropdown(options=df.columns, value= 'GLU', description='caracteristică 1'),
                     column_2= widgets.Dropdown(options=df.columns, value='LTG', description='caracteristică 2'))
```

* Similar cu Machine Learning clasic, insa folosesc o modelare/o forma a modelului mai **generica** (vs dreapta de demarcație, funcția polinomială)
* **Retelele neurale** adanci (deep) sunt o clasa de algorithmi mai puternici, care pot invata reprezentari complexe din date
* **Antrenarea** rețelei neurale = gasirea de parametrii optimi pentru aceste funcții, pentru datasetul invatat

```python
#@markdown Rulați această celulă pentru a vedea performanțele unei rețele neurale simple pentru setul de date considerat.

from sklearn.neural_network import MLPClassifier

def mlp():

  cols = df.columns[:-1]
  samples = df[cols].to_numpy()

  labels = df['diagnostic'].tolist()
  labels = [float(label == 'bolnav') for label in labels]
  labels = np.array(labels)

  clf = MLPClassifier(hidden_layer_sizes=(100,), random_state=1, max_iter=500, batch_size=10, verbose=False, tol=1e-3).fit(samples, labels)

  pred_labels = clf.predict(samples)
  #print(f'Număr parametrii: {polynomial_samples.shape[1]}')

  # calculăm diagnosticul personal
  samples_personal = personal_df[cols].to_numpy()
  pred_labels_personal = clf.predict(samples_personal)

  if pred_labels_personal == 1:
    diag = 'bolnav'
  else:
    diag = 'sanatos'

  print('-------------------------------------------------------------')
  print('Diagnosticul dumneavoastră este: %s'%diag)
  print('-------------------------------------------------------------')


  p_sick = np.sum(labels * pred_labels) / np.sum(labels)
  p_healthy = np.sum((1-labels) * (1-pred_labels)) / np.sum(1-labels)
  acc = (np.sum(labels * pred_labels) + np.sum((1-labels) * (1-pred_labels)))/ len(df)
  print("%4.2f %% din pacienții cu eticheta \'bolnav' au fost corect identificați"%(p_sick*100))
  print('%4.2f %% din pacienții cu eticheta \'sanatos\' au fost corect identificați'%(p_healthy*100))
  print('%4.2f %% din pacienți au fost diagnosticați corect'%(acc*100))
  print('-------------------------------------------------------------')

mlp()
```

Diferit fata de modelele clasice de Machine Learning, Deep Learning vine in plus cu:

<font color='#1FED24'>**Avantaje**</font>:
* performanta foarte buna
* actualitate (toate modelele de Machine Learning depind in mare masura de date si de calitatea acestora. ideal e sa continui sa culegi seturi de date calitative si sa iti imbunatatesti continuu/reantrenezi modelul)

<font color='#ED1F24'>**Dezavantaje**</font>:
* scump computational (in functie de problema, ar putea fi nevoie de hardware dedicat - procesoare grafice/GPUs - pentru a antrena un astfel de model)
* nevoia de expertiza in Deep Learning (modele mai dificil de antrenat)

## Ex3. AnimGANv2 - Transformare stilului in anime

```python
# @markdown Rulați această celulă pentru a testa cum functioneaza AnimGANv2 pornind de la o imagine data ca input
# Code from [AnimGANv2 github](https://github.com/bryandlee/animegan2-pytorch). [Test it on huggingface](https://huggingface.co/spaces/akhaliq/AnimeGANv2).

from io import BytesIO
import torch
from PIL import Image

import ipywidgets as widgets
import IPython.display as display
from google.colab import files

device = "cuda" if torch.cuda.is_available() else "cpu"
model = torch.hub.load("bryandlee/animegan2-pytorch:main", "generator", device=device).eval()
face2paint = torch.hub.load("bryandlee/animegan2-pytorch:main", "face2paint", device=device)
image_format = "png" #@param ["jpeg", "png"]

button = widgets.Button(description="Start")
output = widgets.Output()

print("Incarcati o poza .png")
def run(b):
    button.disabled = True

    with output:
        display.clear_output()

    uploaded = files.upload()

    for fname in uploaded:
        bytes_in = uploaded[fname]

        im_in = Image.open(BytesIO(bytes_in)).convert("RGB")
        im_out = face2paint(model, im_in, side_by_side=False)
        buffer_out = BytesIO()
        im_out.save(buffer_out, format=image_format)
        bytes_out = buffer_out.getvalue()
        wi1 = widgets.Image(value=bytes_in, format=image_format)
        wi2 = widgets.Image(value=bytes_out, format=image_format)
        wi1.layout.max_width = '500px'
        wi1.layout.max_height = '500px'
        wi2.layout.max_width = '500px'
        wi2.layout.max_height = '500px'

        ## Side by side thanks to HBox widgets
        sidebyside = widgets.HBox([wi1, wi2])
        ## Finally, show.
        with output:
            display.display(sidebyside)

    button.disabled = False

button.on_click(run)
display.display(button, output)
```
