```python
import folium
```

```python
# Definirea locaților
locations = {
    'Ploiești': (44.9333, 26.0333),
    'Craiova': (44.3193, 23.8006),
    'Bacău': (46.5688, 26.9160),
    'Râmnicu Vâlcea': (45.0997, 24.3693),
    'Sibiu': (45.7936, 24.1213),
    'Timișoara': (45.7606, 21.2267),
    'Piteşti': (44.8565, 24.8692),
    'Arad': (46.1866, 21.3123),
    'Cluj-Napoca': (46.7704, 23.5914),
    'Bucharest': (44.4396, 26.0963),
    'Baia Mare': (47.6567, 23.5850),
    'Oradea': (47.0465, 21.9189),
    'Galaţi': (45.4353, 28.0080),
    'Brașov': (45.6579, 25.6011),
    'Iași': (47.1635, 27.5825),
    'Constanţa': (44.1598, 28.6348),
    'Brăila': (45.2652, 27.9595)
}
```

```python
# Crearea hărții centrale în România
m = folium.Map(location=[45.9432, 24.9668], zoom_start=7)
```

```python
# Adăugarea locațiilor pe hartă
for city, coord in locations.items():
    folium.Marker(location=coord, popup=city).add_to(m)
```
